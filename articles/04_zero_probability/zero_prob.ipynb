Consider a random event $X$ having probability $p(X)=0$. Does it mean $X$ cannot occur, i.e. is impossible?

# Probability space

Consider the set $\Omega$ containing all possible events, or outcomes of a random process. Let $\mathcal{F}$ be a subset of $\Omega$ containing events that meet a certain criteria, which we can also call a *family* of events. The probability of a family of events $\mathcal{F}$ is the mapping $P(\mathcal{F}) \to \mathbb{R}$, which describes the applicable proability distribution between different events $\mathcal{F}$.

The triple $(\Omega, \mathcal{F}, P)$ defines the probability space.

## Discrete random variables
A typical example is a six-sided dice, where the outcomes of one toss are in the set $\Omega = \{1, 2, \dots, 6\}$. $\mathcal{F}$ are all possible subsets of $\Omega$. 

Let the random variable $X$ describe the outcome of a dice throw, i.e. $X \in \Omega$. We can define the subsets $\mathcal{F}$ based on desired values of $X$, i.e. for $X=2$ or for $X$ to be even, as an example.

The probability of a familiy of events $\mathcal{F}$, i.e. $P(\mathcal{F})$ - assuming uniform probabilities - is the mapping

$$P(\mathcal{F}) = \frac{\text{\# of events in } \mathcal{F}}{\text{Total events in } \Omega}$$

So the probability of a singe value of $X$, i.e. $\mathcal{F} = \{X | X \in \Omega, X=2\} = \{2\}$ is $P(\mathcal{F}) = P(X=2) = 1/6$ and is $3/6=1/2$ to throw an even or odd value.

## Continuous random variables
For the dice, the random variable $X$ is discrete, i.e. it can only take certain specific values. We can also consider a continous random variable.

Let $x$ take values between 0 and 1, both included. Then we can define 

$$\Omega = [0, 1] = \{x | 0 \leq x \leq 1\}.$$

This outcome set has infinitely many elmements, i.e. there are infinitely many real numbers between 0 and 1. Consequently, a uniform probability mapping $P$ is undefined due to the denominator being infinite. We can still assign a value to the probability of choosing a single real number $a$ from a subset $\Omega_N$ with $N$ elements, which contains $a$, and let $\Omega_N \to \Omega$

$$P(X=a) = \lim_{N \to \infty} \frac{1}{N} = 0.$$

If the subsets $\Omega_N$ do not contain a, e.g. choosing $a=1.1$, the probability is trivial as

$$P(X=a) = \lim_{N \to \infty} \frac{0}{N} = 0.$$

So here we have two possibilities for drawing a random variable $X \in \Omega$ which both have zero probability - are they both impossible?
