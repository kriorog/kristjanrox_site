Consider a random event $X$ having probability $p(X)=0$. Does it mean $X$ cannot occur, i.e. is impossible?

# An informal approach
Let's consider a random process which assigns values to the random variable $X$. Let $\Omega$ be the set of all possible outcomes of $X$, i.e. we draw $X$ from $\Omega$ or that $X \in \Omega$. Now, $\Omega$ can be a finite with only $N$ elements or an infinite set.

Given this information, we can't say what the probability of $X$ is, i.e. $P(X)$. The probaility function is one that we have assigned to the random process, which we are confident will describe the relative frequency of different $X$ values from repeated trials of the random process. It is therefore dependent on what elements are in $\Omega$.

## A finite number of possible outcomes
Say that $\Omega$ contains all natural numbers between 1 and 6, i.e. $\Omega = \{1, 2, \dots , 6\}$. This could describe the possible outcomes of a single standard die. The random variable $X$ then describes the value obtained in each trial, i.e. each die cast. The probability of $X$ taking a specific value in $\Omega$ depends on the die itself and how it is cast. 

Let's assume the die to be  *perfect* and the cast process to be fair, then one value in $\Omega$ will not occur more often than any other value in repeated trials. For such a process, the frequency of all values approach the same value as the number of trials increases. For such a result, a *discrete uniform probability distribution* would be an appropriate probability function for this process. We would then say that the die value in each cast follows a uniform distribution, i.e. $X \sim \mathcal{U}(N_{\Omega})$ and the probability of obtaining a value $X=a$, where $a \in \Omega$, is $P(X=a)=1/N_{\Omega} = 1/6$.

This is the same interpretation as the classical definition of probability a la Bernouilli and Laplace:

:::{.callout-note title="Classical definition of probability"}
The probability of an event is the ratio of the number of cases favorable to it, to the number of all cases possible when nothing leads us to expect that any one of these cases should occur more than any other, which renders them, for us, equally possible.
:::

Back to the question...note that each possible outcome of $X$ is equally probable, i.e. $P(X=a)=1/6$ for all $a \in \Omega$. So only values not in $\Omega$ will have probability zero, i.e. $P(X=a)=0$ only if $a \notin \Omega$. That is not an interesting results, an I would say it to be trivial. The only such results which one could possibly argue to be theoretically possible is for the die not to return any element in $\Omega$, i.e. to return the empty set. That would describe the case where the die ends up balanced on an edge, without any particular side facing up. However, such an outcome has most likely never been observed and is accurately assigned a zero probability by the uniform distribution, i.e. $P(X=\emptyset)=0$. However, an outcome of values $X>6$ are absurd - and have $P(X>6)=0$ so each value also has zero probability since $P(X>6) = 1 - P(X \leq 6) = 1-1 =0$ - so this result is indeed impossible.

Therefore, for the discrete random variable case, i.e. $X$ is discrete, with finite possible outcomes in $\Omega$, the only outcomes with zero probability are those which are impossible by definition.

## What if $N_{\Omega} \to \infty$?
Consider now that the size of the finite outcome set $\Omega$ increases towards $\infty$. We can do that in essentially two ways, either $\Omega$ includes more and more of until $\Omega$ equals the set of natural numbers, i.e. $\Omega = \mathbb{N}$, or it includes more values between $1$ and $6$ such that $\Omega \to [1, 6]$. 

A key difference in $\Omega$ between these two options, is that the first case where $\Omega \to \mathbb{N}$ is that $\Omega$ remains *countably infinite*, i.e. there exists a one-to-one mapping between each element of $\Omega$ and the natural numbers. In the latter case, where $\Omega \to [1, 6]$, is that $\Omega$ becomes *uncountably infinite*.

In both cases, the number of possible outcomes approaches infinity with increasing size of $\Omega$. As the uniform probability assumption still holds, each element in the set is equally likely to be obtained in each trial. Therefore, we can state that
$$
P(X=a) = \lim_{N_{\Omega} \to \infty} \frac{1}{N_{\Omega}} = 0.
$$

However, if $N_{\Omega}$ remains finite, although approaching infinity the probability $1/N_{\Omega}$ is defined and finite, although possibly very small or infinitesimal, but not zero. When $\Omega$ has become either $\mathbb{N}$ or $[1,6]$ the discrete uniform distribution is no longer defined.

However, when $\Omega = [1,6]$ the outcome set is a continous range of reals. For such a problem, we can use a continous uniform distribution. We can use a probability density function $f_X(x) = 1/(6-1)=1/5$ where the probability of obtaining a value $X$ such that $1 \leq X \leq a$ is 
$$
P(1 \leq X \leq a) = \int_{1}^{a}f_X(a)dx
$$

Note here that the probability of obtaining $X=a$ is 
$$
P(X = a) = P(a \leq X \leq a) = \int_{a}^{a} f_X(a)dx = F_X(a) - F_X(a) = 0.
$$

So the probability of obtaining exactly a specific remains zero, but the probability of obtaining a value in an interval is non-zero, with $P(1 \leq X \leq 6) = 1$. This matches our previous conclusion, that as the number of possible outcomes increases to infinity the probability of a specific value approaches zero.

Applying the same logic to the case when $\Omega = \mathbb{N}$ is not as well defined, as it now has become a countably infinite set of discrete values. A discrete uniform distribution cannot be applied on such a set, since the probability of each value would be zero while the sum of the probability of each value must sum to unity. A density function cannot be defined as well since the possible values are not continuous. However, for the sake of this argument, we can use the limit approach as $\Omega \to \mathcal{N}$ and say that $P(X=a) = 1/N_{\Omega} \to 0$ for all $a \in \Omega$.

Regardless - some value must be obtained in each trial. So why can't, say, $X=4$ regardless of $\Omega = \mathcal{N}$ or $\Omega = [1,6]$? It can, absolutely, the problem is that it's just extremely improbable - yet not impossible - compared to all possible outcomes. We just have a problem assigning non-zero probabilities to values from discrete and countably infinite outcome sets. Therefore, an outcome $X=4$ is possible, even though $P(X=4)=0$.

## Conclusion
Given the preceeding informal discussion, we can argue that an event with zero probability does not have to be impossible - but an impossible event will have zero probability. 

# A more formal approach - using probability spaces

Consider the set $\Omega$ containing all possible events, or outcomes of a random process. Let $\mathcal{F}$ be a subset of $\Omega$ containing events that meet a certain criteria, which we can also call the event space. The probability function $P$ is a set function that assigns each event $E \in \mathcal{F}$ its probability $P(E): \mathcal{F} \subseteq \Omega \to [0, 1]$.

The triple $(\Omega, \mathcal{F}, P)$ defines the probability space, which satisfies the [Kolmogorov aixioms](https://en.wikipedia.org/wiki/Probability_axioms).

## Discrete random variables
A typical example is a six-sided dice, where the outcomes of one toss are in the set $\Omega = \{1, 2, \dots, 6\}$.

Let the random variable $X$ describe the outcome of a die throw, i.e. $X \in \Omega$. We can define the event space $\mathcal{F}$ based on desired values of $X$, i.e. for $X=2$ or for $X$ to be even, as an example.

The probability of the entire event space $\mathcal{F}$, i.e. $P(\mathcal{F})$ - assuming uniform probabilities - is the mapping

$$P(\mathcal{F}) = \frac{\text{\# of events in } \mathcal{F}}{\text{Total events in } \Omega} = \sum_{j=1}^{N_{\mathcal{F}}} P(E_j)$$

where $N_{\mathcal{F}}$ is the size of the event space.

So the probability of a singe value of $X$, i.e. 

$$\mathcal{F} = \{X | X \in \Omega, X=2\} = \{2\}$$

is $P(\mathcal{F}) = P(X=2) = 1/6$ and is $3/6=1/2$ to throw an even or odd value.

## Continuous random variables
For the dice, the random variable $X$ is discrete, i.e. it can only take certain specific values. We can also consider a continous random variable.

Let $x$ take values between 0 and 1, both included. Then we can define 

$$\Omega = [0, 1] = \{x | 0 \leq x \leq 1\}.$$

This outcome set has infinitely many elmements, i.e. there are infinitely many real numbers between 0 and 1. Consequently, a uniform probability mapping $P$ is undefined due to the denominator being infinite. We can still assign a value to the probability of choosing a single real number $a$ from a subset $\Omega_N$ with $N$ elements, which contains $a$, and let $\Omega_N \to \Omega$

$$P(X=a) = \lim_{N \to \infty} \frac{1}{N} = 0.$$

If the subsets $\Omega_N$ do not contain a, e.g. choosing $a=1.1$, the probability is trivial as

$$P(X=a) = \lim_{N \to \infty} \frac{0}{N} = 0.$$

So here we have two possibilities for drawing a random variable $X \in \Omega$ which both have zero probability - are they both impossible?

# Impossibility
There is only one way for the event space $\mathcal{F}$ to be impossible. That is for at least one event in $\mathcal{F}$ to not be in $\Omega$. 

For example, throwing a six-sided dice and getting 7 is impossible, which would have zero probability and is impossible.

However, a fully reasonable outcome with zero probability is not impossible, but rather *improbable*. Drawing $a=0.5$ from the real interval $[0, 1]$ under uniform probabilities is perferctly reasonable, but extremely improbable considering the infinite number of possible draws. However, drawing $a=1.1$ is impossible, similar to the discrete dice above.

Therefore, an event with zero probability does not imply impossibility. However, the reverse is true - an impossible event has zero probability.
